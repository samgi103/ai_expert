## 한글 영화 데이터 처리

```python
# 필요한 라이브러리 불러오기
import pandas as pd
import urllib.request
```

```python
# 영화 리뷰 정보 다운로드하기
urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings_train.txt", filename="ratings_train.txt")
urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings_test.txt", filename="ratings_test.txt")
```

```python
# Pandas로 데이터 불러오기
train_data = pd.read_table('ratings_train.txt')
test_data = pd.read_table('ratings_test.txt')
```

```python
train_data.iloc[0]
```

```python
# 데이터 사이즈 확인
train_data.shape, test_data.shape

# 보통
# train : 4/5, test : 1/5
# cross-validation 1:3
# train : 3/5, val : 1/5
```

```python
# 데이터 확인
train_data.head()
```

```python
# 2. Stopwords 및 특수문자 필터링을 위한 준비

comment_words = ''
stopwords = set(stopwords)

train_tokenized_text = []

# 3. 문장 데이터를 단어화하기
for val in train_data["document"]:

  # 문장을 string으로 만들기
  val = str(val)

  # 문장을 쪼개기
  tokens = val.split()
  # tokens = Okt.morphs(val)

  train_tokenized_text.append(tokens)

test_tokenized_text = []
for val in test_data["document"]:

  # 문장을 string으로 만들기
  val = str(val)

  # 문장을 쪼개기
  tokens = val.split()
  # tokens = Okt.morphs(val)

  test_tokenized_text.append(tokens)
```

```python
def word_feats(words):
    return dict([(word, True) for word in words if word not in stopwords])
train_feats = list(
    zip(
        [word_feats(doc) for doc in train_tokenized_text],
        train_data["label"].values.tolist()
        )
    )
test_feats = list(
    zip(
        [word_feats(doc) for doc in test_tokenized_text],
        test_data["label"].values.tolist()
        )
    )
```

```python
from nltk import NaiveBayesClassifier
classifier = NaiveBayesClassifier.train(train_feats)
```

```python
print('Accuracy: ', nltk.classify.util.accuracy(classifier, test_feats))
```

```python
# Classifier 1
import collections
refsets = collections.defaultdict(set) # For what this is: https://stackoverflow.com/questions/5900578/how-does-collections-defaultdict-work
testsets = collections.defaultdict(set)

for i, (feats, label) in enumerate(test_feats):
    refsets[label].add(i)
    observed = classifier.classify(feats)
    testsets[observed].add(i)
```

```python
# 1 => positive

TP = len(refsets[1].intersection(testsets[1]))
FP = len(refsets[0].intersection(testsets[1]))
FN = len(refsets[1].intersection(testsets[0]))
TN = len(refsets[0].intersection(testsets[0]))

TP, FP, FN, TN
```

```python
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1 = 2 * precision * recall / (precision + recall)
precision, recall, f1
```

```python
from nltk.metrics.scores import (precision, recall, f_measure)

# We can proceed to print the metrics for each classifier. However, we cannot get the accuracy in this manner because nltk.metrics.scores.accuracy(reference, test) works by comparing test[i] == reference[i] and our reference and test are not formatted in a way that allows for this. It's the same for the confusion matrix.
args = (
    round(precision(refsets[1], testsets[1]), 2),
    round(recall(refsets[1], testsets[1]), 2),
    round(f_measure(refsets[1], testsets[1]), 2),
)


print('''
CLASSIFY Result
------------
Positive precision: {0}
Positive recall: {1}
Positive F1 score: {2}

'''.format(*args))
```
